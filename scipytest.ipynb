```python
import scipy.constants as C
import matplotlib.pyplot as plt
import numpy as np
print(C.pi)
print(C.g)
```

```python
import scipy.special as spl
x = np.linspace(0, 20, 500)
%matplotlib inline
for i in range(3):
    y = spl.jv(i, x)
    plt.plot(x, y, '-', label="J%d" %i)
```

```python
#积分计算
from scipy import integrate
Y = lambda x:x**2+3
print(integrate.quad(Y, -2, 4))
```
